# BEGINNER GUIDE TO WordCloud Python

## Imports

```python
import numpy as np
from PIL import Image
from collections import Counter

import matplotlib.pyplot as plt
import wordcloud 

import json
from nltk.corpus import stopwords
from skimage.io import imsave
```

## Read the Document

```python
document = "Path-to-/scrap.txt"
```

```python
words = []
with open(document, "r") as input_file:
    for line in input_file:
        line = line.lower()
        line = line.strip().split()
        words += line  
print("Document Loaded")
```

## Remove StopWords


```python
filtered_words = []
with open('Path-to-/stopwords-fr.json',encoding="utf-8") as stop:
    mystopwords = json.load(stop)
stopeng=stopwords.words("english")
stopfr=stopwords.words("french")
for word in words:
    if not word in mystopwords and not word in stopeng and not word in stopfr and word.isalpha():
        filtered_words.append(word)
print("StopWords Removed")
```

## Count words

```python
word_counts = Counter(filtered_words)
```

## Read the image for the WordCloud

```python
imgfile = "Path-to-/image.jpg"
bgimg = np.array(Image.open(imgfile))
print("Image Loaded")
```

## Draw  and Save the WordCloud

```python
cloud = wordcloud.WordCloud(background_color="white", mask = bgimg, collocations = False)
word_cloud = cloud.generate_from_frequencies(word_counts)
cloud = word_cloud.to_file('WordCloud.png')#Save the Image
print("WordCloud saved to Disk")
plt.imshow(word_cloud)#Show the Image
plt.axis("off")
plt.show()
```
